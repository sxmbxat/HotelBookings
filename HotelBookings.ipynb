```python
resort_hot= len(hotel_booking[hotel_booking['hotel']=='Resort Hotel'])
city_hot=len(hotel_booking[hotel_booking['hotel']=='City Hotel'])

hotel_types = ['Resort Hotel', 'City Hotel']
lengths = [resort_hot, city_hot]

plt.bar(hotel_types, lengths)
plt.xlabel('Hotel Type')
plt.ylabel('Number of Guest')
plt.title('Number of Guest for Each Hotel Type')
plt.show()
```

```python
country_count = hotel_booking[hotel_booking['is_canceled'] == 0]['country'].value_counts().reset_index()
country_count.columns = ['country', 'No of guests']
country_count
```

```python
data_resort = hotel_booking[(hotel_booking['hotel'] == 'Resort Hotel') & (hotel_booking['is_canceled'] == 0)]
data_city = hotel_booking[(hotel_booking['hotel'] == 'City Hotel') & (hotel_booking['is_canceled'] == 0)]

resort_hotel = data_resort.groupby(['arrival_date_month'])['adr'].mean().reset_index()
resort_hotel
```

```python
city_hotel = data_city.groupby(['arrival_date_month'])['adr'].mean().reset_index()
city_hotel
```
